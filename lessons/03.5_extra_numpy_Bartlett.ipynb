````python
new_points = np.arange(0, 500, 1)
even_points = new_points[new_points % 2 == 0] #Since a point is even iff it has an even index, this gets the even indexed points
plt.plot(even_points, np.sin(2*np.pi*even_points))
#This plot look very different because it is graphing sin from 0 to 1000pi rather than from 0 to 2 pi
#Theoretically, Sin(2*pi*n) whould always be 0 if n is an integer, and this appears to hold true for small n in the graph,
#However, later in the graph the sin wave gives values further and further from zero, likely due to floating point precision
#error. Also, looking at the next exercise, it seems like this was supposed to be 500 points from 0 to 1, so I did that on the
#next cell, however I will leave this as I think it is an interesting graph.
````
